```python
df_severity = sorted(list(data.Severity.unique()))
Severity = pd.DataFrame(df_severity, columns=['Severity'])
Severity['Id'] = [x for x in range(len(df_severity))]
Severity
```

```python
# Converting Object Ride Type to Int
def transform_severity(severity):
    if severity == 1:
        return 0
    elif severity == 2:
        return 1
    elif severity == 3:
        return 2
    elif severity == 4:
        return 3
    else:
        return -1
data['Severity'] = data['Severity'].apply(transform_severity)
```

```python
# State Table
df_state = list(data.State.unique())
state = pd.DataFrame(df_state, columns=['state_name'])
state['state_id'] = [x for x in range(len(df_state))]
state
```

```python
def transform_state(st):
  if st in state['state_name'].unique():
    matched = list(state[state['state_name'] == st]['state_id'])
    return matched[0]
data['State'] = data['State'].apply(transform_state)
```

```python
data['State']
```

```python
# Airport Table
df_airports = list(data.Airport_Code.unique())
airports = pd.DataFrame(df_airports, columns=['airport_code'])
airports['airport_id'] = [x for x in range(len(df_airports))]
airports
```

```python
def transform_airports(ap_code):
  if ap_code in airports['airport_code'].unique():
    matched = list(airports[airports['airport_code'] == ap_code]['airport_id'])
    return matched[0]
data['Airport_Code'] = data['Airport_Code'].apply(transform_airports)
data['Airport_Code']
```
